**datetime 연습**

```python
import datetime
```

```python
# datetime.now()를 이용하여 현재 시간의 시계열 데이터 생성
now = datetime.datetime.now()
print(now)
```

```python
# datetime()를 이용하여 시계열 데이터 생성
sp_date = datetime.datetime(2002,2,17,12,30,45)
print(sp_date)
```

```python
type(now)
```

```python
type(sp_date)
```

```python
print(now.year) #년도
print(now.month) #월
print(now.day) #일
print(now.hour) #시
print(now.minute) #분
print(now.second) #초
print(now.weekday()) #요일(0~6) -> 0 : 월요일 ~ 6 : 일요일
```

```python
datetime.date.today() #오늘 날짜
```

```python
# strftime()을 이용하여 시계열 데이터 -> 문자열 데이터로 변환
after_strftime_1 = now.strftime('%Y-%m-%d %H:%M:%S')
print(after_strftime_1)
type(after_strftime_1) # 문자열 데이터로 변환되었음을 확인

after_strftime_2 = now.strftime('%Y-%m-%d')
print(after_strftime_2)
after_strftime_3 = sp_date.strftime('%Y-%m')
print(after_strftime_3)
```

```python
# strptime()을 이용하여 시계열 데이터 생성
after_strptime = datetime.datetime.strptime('2000-03-20 13:03:23', '%Y-%m-%d %H:%M:%S')
print(after_strptime)
type(after_strptime) # 시계열 데이터가 생성됨을 확인
```

**timedelta 연습**

```python
# timedelta()를 이용하여 두 날짜/시간의 차이를 계산
after_timedelta = datetime.timedelta(days=100, hours=3)
print(after_timedelta)
```

```python
print(now + after_timedelta) # 미래 데이터 구하기
print(now - after_timedelta) # 과거 데이터 구하기
```

```python
# relativedelta()를 이용하여 편리하게 날짜/시간 차이 계산 (ex. 3개월 전 or 1년 후)
from dateutil.relativedelta import relativedelta

print(now +relativedelta(months=12))
print(now +relativedelta(years=1))
print(now -relativedelta(days=1))
```

**Pandas 문법 연습**

- freq : D(day), y(year), m(month), w(week)

```python
import pandas as pd
```

```python
# 3개(periods)의 Day 간격(freq) 시계열 데이터 생성
after_date_range_1 = pd.date_range('2020-01-01', periods=3, freq='D')
print(after_date_range_1)
```

```python
# start~end 까지의 month 간격(freq) 시계열 데이터 생성
after_date_range_2 = pd.date_range(start='2020-01-01', end='2021-01-01', freq='m')
print(after_date_range_2)
```

```python
#calendar을 이용하여 달력 생성
import calendar

cal_7 = calendar.month(2024,7)
print(cal_7)
```

# **시계열 데이터 분석해보기**
분석할 데이터 : crime.csv => 특정 지역에서 발생한 범죄와 교통사고에 대한 정보

**Unnamed: 0** : 인덱스

**OFFENSE_TYPE_ID**: 범죄나 사건의 구체적인 유형

**OFFENSE_CATEGORY_ID**: 범죄나 사건의 카테고리

**REPORTED_DATE**: 사건이 보고된 날짜와 시간

**GEO_LON**: 사건이 발생한 장소의 경도(longitude)

**GEO_LAT**: 사건이 발생한 장소의 위도(latitude)

**NEIGHBORHOOD_ID**: 사건이 발생한 지역의 고유 ID

**IS_CRIME**: 해당 사건이 범죄인지 여부를 나타내는 이진 값(0 또는 1)

**IS_TRAFFIC**: 해당 사건이 교통사고인지 여부를 나타내는 이진 값(0 또는 1)

```python
df = pd.read_csv('crime.csv')
df.info() # REPORTED_DATE의 자료형이 object이므로 시계열데이터로 변환해야 한다 !
```

```python
# 1. pd.to_datetime()을 이용하여 문자열 -> 시계열 데이터로 변환
df_time = df.copy()
df_time['REPORTED_DATE'] = pd.to_datetime(df_time['REPORTED_DATE'])
df_time.info()
```

```python
# 2. 시계열 데이터를 분석하기 위해 set_index()를 이용하여 Datetime형을 인덱스로 설정

df_time=df_time.set_index('REPORTED_DATE')
```

```python
# 2014년도에 대한 범죄/교통사고 정보만을 추출
df_time.loc['2014']
```

```python
# 2014년도 6월 29일에 대한 범죄/교통사고 정보만을 추출
df_time.loc['2014-06-29']
```

```python
# 2014년도 10월에 대한 범죄/교통사고 정보만을 추출
df_time.loc['Oct 2014']
```

```python
# 데이터 정렬
df_time.sort_index()
```

```python
# between_time()를 이용하여 시간대별(01:00 ~ 02:00) 교통사고/범죄 정보만을 추출
df_time.between_time('01:00','02:00')
```

```python
# at_time()를 이용하여 시간대별(18:00) 교통사고/범죄 정보만을 추출
df_time.at_time('18:00')
```

```python
# resample()를 이용하여 특정 기간별 컬럼을 추출

df_crime=df_time[['IS_CRIME']] # IS_CRIME 컬럼만을 추출
df_crime.resample('M').sum() # 월별 범죄건수
```

```python
df_crime.resample('Y').sum() # 년도별 범죄건수
```

```python
df_crime.resample('W').sum() # 주별 범죄건수
```

```python
df_time['IS_CRIME'].resample('M').sum() # 월별 범죄건수
```

```python
df_time[['IS_CRIME','IS_TRAFFIC']].resample('Y').sum() # 년도별 범죄건수, 교통사고수
```

```python
df_time
```

```python
# shift()를 이용하여 데이터를 지정한 기간만큼 이동
df_time.shift(3)
```

```python
# rolling()을 이용하여 이동 평균,이동 합계 등의 이동 통계량을 계산
df_time['IS_CRIME'].rolling(window= 30).mean() # 범죄건수의 30일 이동 평균
```

```python
df_time['IS_CRIME'].rolling(window= 30).sum() # 범죄건수의 30일 이동 합계
```
