**파생변수 만들기**

1. ['새로운 칼럼명'] = 0
2. assign(변수명=조건식, 조건에 맞는 값, 그외 값)

```python
tt['sex_onehot']=0 # 새로운 파생변수(칼럼) 생성 (default 값이 0)
tt
```

```python
# 성별이 남성이면 0을 여성이면 1을 나타내는 파생변수 sex_onthot 완성하기

for i in range(len(tt)) :
  if tt.sex.loc[i] == 'male' :
    tt['sex_onehot'].loc[i] = 0
  else :
    tt['sex_onehot'].loc[i] = 1

tt
```

```python
print(tt.sex_onehot.value_counts()) # sex_onthot 칼럼의 개수 카운트
```

```python
print(tt.sex.value_counts()) # sex 칼럼의 개수 카운트
```

**Group by**
: 그룹화하는 함수

```python
grouped = tt.groupby('pclass') # pclass 기준으로 그룹화한다.(결과는 grouped에 저장)
```

```python
grouped
```

```python
for i,j in grouped :
  print(i)
  print(j.head(10))
```

```python
# 그룹화한 결과(grouped)에서 원하는 소그룹만 출력하기
grouped.get_group(2)
```

```python
# 2가지 이상의 기준을 가지고 그룹화하기 -> 리스트 이용
groups = tt.groupby(['pclass','survived'])
```

```python
groups
```

```python
for i,j in groups :
  print(i)
  print(j.head(10))
```

```python
groups.mean() # 평균구하기
```

```python
groups.std() # 표준편차구하기
```

**groupby 에서 통계치 살펴보기**

: agg함수를 이용한다

groupby('칼럼').agg(변수명 = ('기준칼럼1','통계치'),...)

```python
tt.groupby('pclass').agg(survived_mean = ('survived','mean'))
```

```python
tt.groupby('pclass').agg(fare_sum = ('fare','sum'))
```

```python
tt.groupby('pclass').agg(survived_sum = ('survived','sum'),
                         survived_mean = ('survived','mean'),
                         survived_count = ('survived','count'))
```

```python
def min_max(x):
  return x.max() - x.min() # 함수 정의
```

```python
grouped.agg(min_max)
```

```python
grouped.agg({'survived' : ['sum', 'mean', 'std']})
```

**데이터 병합하기**

1. concat : 덩어리와 덩어리가 붙음 (그대로)

pd.concat([컬럼1,컬럼2], axis =0/1)

axis = 0 은 위/아래로 붙음, axis = 1 은 옆으로 붙음

2. merge : 덩어리와 덩어리가 병합하여 하나의 새로운 덩어리 생성

pd.merge(왼쪽컬럼(기준), 오른쪽컬럼(그다음), how='병합방법', on='공통기준컬럼')

```python
one = pd.DataFrame({'학회원번호' : [1,2,3,4,5],
                    '학회원분반' : ['Python', 'SQL', 'R', '데이터분석입문','데이터분석전처리']})
```

```python
two = pd.DataFrame({'학회원번호' : [1,2,3,4,5],
                    '상점현황': [5,5,10,0,3]})
```

```python
pd.merge(one,two, how = 'inner', on = '학회원번호')
```

```python
pd.concat([one,two])
```

```python
pd.concat([one,two], axis = 1)
```
